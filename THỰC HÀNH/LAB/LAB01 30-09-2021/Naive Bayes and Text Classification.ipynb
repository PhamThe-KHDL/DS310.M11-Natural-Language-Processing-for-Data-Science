```python
train = [
    ["just plain boring", "-"],
    ["entirely predictable and lacks energy", "-"],
    ["no surprises and very few laughs", "-"],
    ["very powerful", "+"],
    ["the most fun film of the summer", "+"]
]

test = "predictable with no fun"
```

```python
test.split()
```

## Hien thuc bang tay

```python
# Xây dựng tập từ vựng 
vocabulary = []
vocabulary_positive = []
vocabulary_negative = []
y_train = []

for s in train:
    if s[1] == "-":
        vocabulary_negative += s[0].split()
    else:
        vocabulary_positive += s[0].split()
    
    vocabulary += s[0].split()
    y_train.append(s[1])

vocabulary = set(vocabulary)
```

```python
y_train
```

```python
# tính xác suất cho từng nhãn 
test_list = test.split()
p_negative = y_train.count("-") / len(y_train)
p_positive = y_train.count("+") / len(y_train)

p_test_positive = p_positive
p_test_negative = p_negative

# Tính xác suất tổng cho từng từ trong test theo từng nhãn
print(test_list)
for t in test_list:
    if t == "with":
        continue
    p_test_positive *= (vocabulary_positive.count(t)+1) / (len(vocabulary_positive) + len(vocabulary))
    p_test_negative *= (vocabulary_negative.count(t)+1) / (len(vocabulary_negative) + len(vocabulary))

print("P(S|+)={}".format(p_test_positive))
print("P(S|-)={}".format(p_test_negative))

if p_test_positive >= p_test_negative:
    print("The class of \"{}\" is +".format(test))
else:
    print("The class of \"{}\" is -".format(test))   
```

## Hien thuc bang Sklearn

```python
X_train = []
y_train = []

X_test = [test]

for t in train:
    X_train.append(t[0])
    y_train.append(t[1])
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder

vectorizer = CountVectorizer()
le = LabelEncoder()

vectorizer.fit(X_train)
le.fit(y_train)

X_train_encoded = vectorizer.transform(X_train)
y_train_encoded = le.transform(y_train)

X_test_encoded = vectorizer.transform(X_test)
```

```python
X_train_encoded
```

```python
y_train_encoded
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_encoded, y_train_encoded)
```

```python
model.predict_proba(X_test_encoded)
```

```python
y_pred = model.predict(X_test_encoded)

print("The class of \"{}\" is {}".format(test, le.inverse_transform(y_pred)[0]))
```

## N-gram

```python
X_train = []
y_train = []

X_test = [test]

for t in train:
    X_train.append(t[0])
    y_train.append(t[1])
```

```python
# 1-gram
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_1_gram = CountVectorizer()
vectorizer_1_gram.fit(X_train)

vectorizer_1_gram.get_feature_names()
```

```python
X_test_1_gram = vectorizer_1_gram.transform(X_test)
```

```python
print(X_test_1_gram)
```

```python
# 2-gram
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_2_gram = CountVectorizer(analyzer='word', ngram_range=(2, 2))
vectorizer_2_gram.fit(X_train)

vectorizer_2_gram.get_feature_names()
```
